prima di usare fuzzy faccio cleaning su stringhe non ancora matchate: 
- lower
- strip
- remove emoticons/chinese/japanese/korean characters

```python
match_df['twitch_clean'] = ''

j=0
for elem in match_df.steam_name:
    if elem == '':
        match_df.twitch_clean[j] = re.sub(r"[^a-zA-Z0-9 ']",'',match_df.twitch_name[j]).lower().strip()
    else:
        match_df.twitch_clean[j] = ''
    j+=1
```

cleaning della lista steam_games e trasformo in df (associo nome originale a pulito)

```python
steam_df = pd.DataFrame()
steam_df['steam_name'] = steam_games

steam_clean = []
for string in steam_games:
    new_string = re.sub(r"[^a-zA-Z0-9 ']",'',string).lower().strip()
    steam_clean.append(new_string)
steam_df['steam_clean'] = steam_clean
steam_df.head()
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
match_df
```

```python
# cosine distance -> non pesa l'ordine
# levensthein -> numero modifiche per passare da A a B
            # -> d(lev) >= |len(A)-len(B)|
```

```python
matched_results = fuzzymatcher.fuzzy_left_join(match_df,
                                               steam_df,
                                              left_on='twitch_clean',
                                              right_on= 'steam_clean')
```
