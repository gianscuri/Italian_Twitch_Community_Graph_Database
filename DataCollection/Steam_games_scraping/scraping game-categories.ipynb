```python
steam_games = {}
contatore = 0

for i in id_games:
    contatore+=1
    print(contatore, end= "\r")
    
    
    driver.get(f'https://steamdb.info/api/RenderAppHover/?appid={i}')
    html = driver.page_source
    window = BeautifulSoup(html, "lxml")
    
    if 'La pagina non funziona' in html:
        pass
    else:
        steam_games[i] = {}
        #title of game
        steam_games[i]['title'] = window.find('a',class_='hover_title').text

        #info game
        info = []
        elenco = window.find_all('div', class_='hover_body hover_meta')
        for elem in range(len(elenco)):
            info.append(elenco[elem].text)
        steam_games[i]['informations'] = info

        #categories
        categories = soup.find('div',class_='hover_body hover_tag_row').text.strip().split('\n')
        steam_games[i]['categories'] = categories
```
